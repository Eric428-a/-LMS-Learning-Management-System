# EXPLANATORY DATA ANALYSIS

```python
# 1. Import Necessary Libraries
# These lines import essential libraries that we'll use for data analysis and visualization.
```

```python
import pandas as pd  # Pandas is used for data manipulation and analysis.
```

```python
import numpy as np   # NumPy provides support for mathematical operations on arrays.
```

```python
import matplotlib.pyplot as plt  # Matplotlib is a plotting library for creating visualizations.
```

```python
import seaborn as sns  # Seaborn is built on top of Matplotlib and provides a higher-level interface for drawing attractive and informative statistical graphics.
```

```python
# 2. Load the Data
# This line reads our dataset from a CSV (Comma-Separated Values) file named 'data.csv' and stores it in a DataFrame called 'df'.
```

```python
df = pd.read_csv('/content/accidental-deaths-in-usa-monthly.csv')
```

```python
# 3. Inspect the Data
# These lines allow us to examine the first few rows of the dataset, obtain information about the data types and non-null counts of each column, and display summary statistics for numerical columns.
```

```python
df.head()       # Display the first few rows of the dataset.
```

```python
df.info()       # Display information about the dataset, including data types and non-null counts.
```

```python
df.describe()   # Display summary statistics of numerical columns.
```

```python
# 4. Check for Missing Values
# This line checks for missing values in our dataset and displays the count of missing values for each column.
```

```python
df.isnull().sum()
```

```python
# 5. Handle Missing Values
# This line fills missing values in our dataset with the mean of each column, ensuring that our analysis is not affected by missing data.
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# 6. Check for Duplicate Values
# This line checks for duplicate rows in our dataset and displays the count of duplicate rows.
```

```python
df.duplicated().sum()
```

```python
# 7. Remove Duplicate Values
# This line removes duplicate rows from our dataset, ensuring that each observation is unique.
```

```python
df.drop_duplicates(inplace=True)
```

```python
# 8. Visualize the Data
# This loop creates histograms with kernel density estimation curves for each column in our dataset, allowing us to visualize the distribution of data.
```

```python
for column in df.columns:
    sns.histplot(df[column], kde=True)    # Create a histogram with kernel density estimation.
    plt.show()                            # Display the plot.
```

```python
# 9. Check Data Distribution
# This line creates histograms for each numerical column in our dataset, providing insight into the distribution of data.
```

```python
df.hist(bins=30, figsize=(20,15), density=True)  # Create histograms for each numerical column.
plt.show()                                      # Display the plots.
```

```python
# 10. Check Correlations
# This line creates a heatmap to visualize the correlation between different variables in our dataset.
```

```python
sns.heatmap(df.corr(), annot=True)  # Create a heatmap of correlation matrix with annotations.
plt.show()                          # Display the heatmap.
```

```python
# 11. Handle Categorical Variables
# This line converts categorical variables into numerical values using one-hot encoding, which is necessary for some machine learning algorithms to process the data.
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
# 12. Outlier Detection
# This section detects and handles outliers in our dataset using the Interquartile Range (IQR) method.
```

```python
Q1 = df.quantile(0.25)      # Calculate the first quartile (25th percentile).
```

```python
Q3 = df.quantile(0.75)      # Calculate the third quartile (75th percentile).
```

```python
IQR = Q3 - Q1               # Calculate the Interquartile Range (IQR).
```

```python
# Keep only the rows that do not have outliers based on the IQR method.
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# 13. Feature Scaling
# This section scales our features to a standard scale using StandardScaler from scikit-learn.
```

```python
from sklearn.preprocessing import StandardScaler  # Import StandardScaler from scikit-learn.
```

```python
scaler = StandardScaler()
```

```python
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)  # Scale the features and convert back to a DataFrame.
```

```python
# 14. Save Processed Data
# This line saves the final processed data to a new CSV file named 'processed_data.csv'.
```

```python
df_scaled.to_csv('processed_data.csv', index=False)  # Specify index=False to exclude row indices from the CSV file.
```
